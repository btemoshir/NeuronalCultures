```python
pop1 = nest.Create("iaf_psc_alpha", 10, {"I_e": 376.})
pop2 = nest.Create("iaf_psc_alpha", 10)
multimeter = nest.Create("multimeter", 10, {"withtime": True, "record_from": ["V_m"]})
```

```python
nest.Connect(pop1, pop2, "one_to_one", syn_spec={"weight": 20., "delay": 1.})
nest.Connect(multimeter,pop2)
```

```python
d = 1.; Je = 2.; Ke = 20; Ji = -4.; Ki =12
conn_dict_ex = {"rule": "fixed_indegree", "indegree": Ke}
conn_dict_in = {"rule": "fixed_indegree", "indegree": Ki}
syn_dict_ex  = {"delay": d, "weight": Je}
syn_dict_in  = {"delay": d, "weight": Ji}
nest.Connect(epop1, ipop1,conn_dict_ex,syn_dict_ex)
nest.Connect(ipop1,epop1, conn_dict_in, syn_dict_in)
```

```python
conn_dict = {"rule": "fixed_indegree", "indegree": 10}
syn_dict  = {"model": "stdp_synapse", "alpha": 1.0}
nest.Connect(epop1, epop2, conn_dict, syn_dict)
```

```python
neuron = nest.Create("iaf_psc_alpha", params={"tau_minus": 30.0})
```

```python
alpha_min = 0.1 ; alpha_max = 2. ; w_min = 0.5 ; w_max = 5.

syn_dict = {"model": "stdp_synapse", "alpha": {"distribution":"uniform", "low": alpha_min, "high": alpha_max},\
           "weight": {"distribution": "uniform", "low": w_min, "high": w_max}, "delay": 1.}
nest.Connect(epop1, neuron, "all_to_all", syn_dict)
```
