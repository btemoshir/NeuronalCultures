```python
import nest
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import scipy as sc
```

```python
neuron = nest.Create("iaf_psc_alpha",params={"I_e": 376.0})
multimeter = nest.Create("multimeter",params={"withtime":True, "record_from":["V_m"] })
spikedetector = nest.Create("spike_detector",params={"withgid": True, "withtime": True})
```

```python
nest.Connect(multimeter, neuron)
nest.Connect(neuron, spikedetector)
```

```python
noise_ex = nest.Create("poisson_generator")
noise_in = nest.Create("poisson_generator")
nest.SetStatus(noise_ex, {"rate": 80000.0})
nest.SetStatus(noise_in, {"rate": 15000.0})
```

```python
nest.SetStatus(neuron, {"I_e": 0.0})
```

```python
syn_dict_ex = {"weight": 1.2}
syn_dict_in = {"weight": -2.0}
nest.Connect(noise_ex, neuron, syn_spec=syn_dict_ex)
nest.Connect(noise_in, neuron, syn_spec=syn_dict_in)
```

```python
nest.Simulate(1000.0)
```

```python
dmm = nest.GetStatus(multimeter)[0]
Vms = dmm["events"]["V_m"]
ts = dmm["events"]["times"]
plt.figure()
plt.plot(ts,Vms)

dSD = nest.GetStatus(spikedetector,keys="events")[0]
evs = dSD["senders"]
ts = dSD["times"]
plt.figure()
plt.plot(ts,evs,'o')
```

```python
neuron1 = nest.Create("iaf_psc_alpha")
nest.SetStatus(neuron1, {"I_e": 376.0})
neuron2 = nest.Create("iaf_psc_alpha")
multimeter2 = nest.Create("multimeter")
nest.SetStatus(multimeter2, {"withtime":True, "record_from":["V_m"]})
```

```python
nest.Connect(neuron1, neuron2, syn_spec = {"weight":20.0})
nest.Connect(multimeter2, neuron2)
```

```python
nest.Simulate(1000.)
```

```python
dmm0 = nest.GetStatus(multimeter2)[0]
Vms1 = dmm0["events"]["V_m"][::2]
ts1 = dmm0["events"]["times"][::2]
Vms2 = dmm0["events"]["V_m"][1::2]
ts2 = dmm0["events"]["times"][1::2]
plt.figure()
plt.plot(ts1,Vms1,'-')
plt.plot(ts2,Vms2,'-.')
```
